```python
import glob

text_files=glob.glob(r"C:\Users\Manivannan\Desktop\ai\*.txt")

for i in text_files[:20]:
    with open(i,"r")as f:
        d=f.read()
        

print(d)
```

```python
import nltk
#nltk..download()
from nltk.tokenize import word_tokenize
```

```python
AI_tokens=word_tokenize(d)
AI_tokens
```

```python
len(AI_tokens)
```

```python
from nltk.probability import FreqDist
fdist=FreqDist()
```

```python
for word in AI_tokens:
    fdist[word.lower()]+=1
fdist
```

```python
fdist_top5=fdist.most_common(20)
fdist_top5
```
